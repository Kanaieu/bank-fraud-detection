**Karena Tidak ada Missing Value dan Duplicate Value maka tidak ada penganganan tersebut**

#### Encoding Data Kategorikal

````python
# Copy dataframe
df_encoded = df.copy()
````

**Mencopy dataframe ke dataframe baru untuk mengencode fitur, ini dilakukan agar dataset asli aman**

````python
# One-Hot Encoding untuk fitur yang memiliki kategori kecil
ohe_features = ["TransactionType", "Channel", "CustomerOccupation"]
df_encoded = pd.get_dummies(df_encoded, columns=ohe_features, drop_first=True)
````

**Disini untuk fitur dengan kategori yang kecil (disini ada 2/3/4 kategori) menggunakan One-hot encoding namun disini menggunakan drop_first agar tidak ada multikolinearitas yang pada model K-means atau DBSCAN itu dapat berpengaruh jika ada fitur berlebih**

````python
# Frequency Encoding untuk Location
location_counts = df["Location"].value_counts().to_dict()
df_encoded["Location"] = df["Location"].map(location_counts)
````

**Disini location diubah valuenya menjadi frekuensi lokasi, karena jumlah kategori yang banyak tidak bisa menggunakan one-hot karena nantinya akan high pada kardinalitas**

````python
# Standarisasi Fitur Numerikal
numerical_features = ["TransactionAmount", "TransactionDuration", "AccountBalance", "LoginAttempts", "Location", "CustomerAge"]

scaler = StandardScaler()
df_encoded[numerical_features] = scaler.fit_transform(df_encoded[numerical_features])
````

**Fitur numerikal yang dipilih adalah fitur numerik yang bersifat kontinu dan berkemungkinan berpengaruh pada clustering, dan bukan hasil dari encoding. Ini diperlukan karena untuk menyamakan skala agar tidak ada pengaruh yang condong jika menggunakan nilai asli**

````python
df_encoded = df_encoded.drop(columns=["TransactionID", "AccountID", "DeviceID", "MerchantID", "TransactionDate", "IP Address", "PreviousTransactionDate"], errors="ignore")
````

**Disini dihapus data string yang memang kemungkinan tidak dapat digunakan pada saat clustering**

````python
print(df_encoded.dtypes)
````

````python
df_encoded.head()
````

````python
# Mencari Epsilon optimal
X = df_encoded.copy()

nn = NearestNeighbors(n_neighbors=5)
nn.fit(X)
distances, _ = nn.kneighbors(X)

# Ambil jarak ke-5 terdekat dan urutkan
distances = np.sort(distances[:, -1])

# Plot untuk menentukan eps optimal
plt.figure(figsize=(8, 6))
plt.plot(distances)
plt.xlabel("Sorted Data Points")
plt.ylabel("k-NN Distance")
plt.title("Elbow Method for Optimal eps")
plt.show()
````

**Eps optimal berada di 2.0 - 2.5**

#### **Elbow Method KMeans**

````python
# Menentukan jumlah klaster optimal dengan Elbow Method
inertia_scores = []
silhouette_scores = []
k_values = range(2, 11)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(df_encoded)
    
    inertia_scores.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(df_encoded, labels))
````

````python
# Menampilkan elbow plot
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(k_values, inertia_scores, marker='o')
plt.xlabel('Jumlah Klaster')
plt.ylabel('Inertia')
plt.title('Elbow Method')

plt.subplot(1, 2, 2)
plt.plot(k_values, silhouette_scores, marker='o', color='red')
plt.xlabel('Jumlah Klaster')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score')

plt.show()
````

Metode ini membantu kita menemukan jumlah cluster yang memberikan pemisahan terbaik antar kelompok data, sehingga model yang dibangun dapat lebih efektif. Berikut adalah **rekomendasi** tahapannya.

### **DBSCAN**

**Variance Threshold**

````python
# Variance Threshold (Menghapus fitur dengan variansi rendah)
var_thresh = VarianceThreshold(threshold=0.2)
df_selected = var_thresh.fit_transform(X)

selected_features = X.columns[var_thresh.get_support()]
print(f"Jumlah fitur yang dipertahankan setelah Variance Threshold: {len(selected_features)}")
print("Fitur yang dipertahankan:")
print(selected_features)
````

**Penggunaan Variance Threshold ini menghapus fitur yang memiliki variansi dibawah threshold yaitu 0.2. Disini total fitur yang memenuhinya ada 8**

````python
# DBSCAN menggunakan feature selection
dbscan_vt = DBSCAN(eps=2, min_samples=9)
dbscan_vt.fit(df_selected)
df_encoded['DBSCAN_Cluster_VT'] = dbscan_vt.labels_
````

**DBSCAN ini ditraining menggunakan data hasil variance threshold dengan eps berasal dari elbow method namun min_sample didapat dari trial and error**

````python
# Evaluasi silhouette score setelah feature selection
if len(set(dbscan_vt.labels_)) > 1:
    silhouette_fs = silhouette_score(df_selected, dbscan_vt.labels_)
else:
    silhouette_fs = -1

print(f"Silhouette Score setelah feature selection (PCA): {silhouette_fs}")
````

````python
import numpy as np

n_outliers = np.sum(dbscan_vt.labels_ == -1)
n_clusters = len(set(dbscan_vt.labels_)) - (1 if -1 in dbscan_vt.labels_ else 0)

print(f"Jumlah outlier: {n_outliers}")
print(f"Jumlah klaster (tidak termasuk outlier): {n_clusters}")
````

**Silhouette Score yang di dapat ada pada 0.38 dengan jumlah outlier 41 dan jumlah klaster tanpa outlier di 3, ini cukup baik namun bisa ditingkatkan silhouette scorenya**

**PCA**

````python
import numpy as np

pca = PCA(n_components=6)
df_pca = pca.fit_transform(df_selected)

dbscan_fs = DBSCAN(eps=2.14, min_samples=19)
df_pca_clusters = dbscan_fs.fit_predict(df_pca)
df_encoded['DBSCAN_Cluster_PCA'] = dbscan_fs.labels_

if len(set(dbscan_fs.labels_)) > 1:
    silhouette_fs = silhouette_score(df_pca, dbscan_fs.labels_)
else:
    silhouette_fs = -1

print(f"Silhouette Score setelah feature selection (PCA): {silhouette_fs}")

n_outliers = np.sum(dbscan_fs.labels_ == -1)
n_clusters = len(set(dbscan_fs.labels_)) - (1 if -1 in dbscan_fs.labels_ else 0)

print(f"Jumlah outlier: {n_outliers}")
print(f"Jumlah klaster (tidak termasuk outlier): {n_clusters}")
````

**Penggunaan PCA meningkatkan silhouette score ke 0.497 dengan jumlah outlier yang berkurang ke 31, dan jumlah klaster yang tidak termasuk outlier menjadi 2**

````python
df_encoded['DBSCAN_Cluster_PCA'] = dbscan_fs.labels_
````

````python
def inverse_one_hot(df, original_feature, one_hot_columns, dropped_category):
    def get_original_category(row):
        for col in one_hot_columns:
            if row[col] == 1:
                return col.replace(f"{original_feature}_", "")
        return dropped_category

    df[original_feature] = df.apply(get_original_category, axis=1)
    return df

# Inverse One-Hot Encoding untuk fitur yang di-drop
df_encoded = inverse_one_hot(df_encoded, "TransactionType", ["TransactionType_Debit"], "Credit")
df_encoded = inverse_one_hot(df_encoded, "Channel", ["Channel_Branch", "Channel_Online"], "ATM")
df_encoded = inverse_one_hot(df_encoded, "CustomerOccupation", ["CustomerOccupation_Engineer", "CustomerOccupation_Retired", "CustomerOccupation_Student"], "Doctor")

# Hapus kolom One-Hot Encoding
ohe_columns = ["TransactionType_Debit", "Channel_Branch", "Channel_Online", 
               "CustomerOccupation_Engineer", "CustomerOccupation_Retired", "CustomerOccupation_Student"]
df_encoded.drop(columns=ohe_columns, inplace=True)
````

**Melakukan Inverse Encoding dimana untuk kategorikal one-hot karena sebelumnya pada saat encoding melakukan drop-first=true, itu menghapus salah satu kategori dari masing maing fitur kategor (contoh seharusnya debit/credit tetapi setelah encoding hanya fitur debit saja karena jika kolom debit false berarti itu credit). Jadi perlu pengecekan bahwa nilai yang dikembalikan itu yang bernilai 1 jika nilai nya 0 maka akan keluar value dari kategori yang didrop saat encoding pertama**

````python
categorical_features = ['Location', 'TransactionType', 'Channel', 'CustomerOccupation']
cluster_column = 'DBSCAN_Cluster_PCA' 

category_distribution = {}

for feature in categorical_features:
    distribution = df_encoded.groupby([cluster_column, feature]).size().unstack(fill_value=0)
    
    if feature == 'Location':
        total_counts = df_encoded['Location'].value_counts() 
        sorted_locations = total_counts.index 
        distribution = distribution[sorted_locations]
    
    category_distribution[feature] = distribution

# Menampilkan hasil dalam bentuk tabel
for feature, table in category_distribution.items():
    print(f"\n📌 Distribusi {feature} per Cluster:\n")
    display(table)
````
